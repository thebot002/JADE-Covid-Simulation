# Displaying graphs of contamination

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from matplotlib.legend import _get_legend_handles_labels
import numpy as np
```

```python
ts = '20230722_113450'
base_path = '../Results/Simulation_'
entries = os.listdir(base_path + ts)

entries = sorted(entries)
```

```python
container_count = len(entries)

# plt.figure(figsize=(6, 3*container_count))
fig, axes = plt.subplots(nrows=container_count, ncols=1,  figsize=(6, 3*container_count), sharex=True, sharey=True,)

for i, container_file in enumerate(entries):
    pd_ds = pd.read_csv(base_path + ts + "/" + container_file)
    
    ys = pd_ds.to_numpy()
    x = np.arange(ys.shape[0])
    
    ax = axes[i]
    
    ax.set_title(container_file.replace(".csv", "").replace("Controller", "Container"))
    ax.stackplot(x, ys[:,3], ys[:,2], ys[:,1], colors=["grey", "red", "blue"], labels=["Recovered", "Sick", "Healthy"])
    ax.legend(loc = "upper left")

plt.show()
```
